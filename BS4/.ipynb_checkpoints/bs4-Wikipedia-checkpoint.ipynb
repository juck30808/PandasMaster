```python
import requests
from bs4 import BeautifulSoup
url ='https://en.wikipedia.org/wiki/Programming_languages_used_in_most_popular_websites'
    
def get_all_href(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    results = soup.select('.unsortable')
    for item in results:
        a_item = item.select_one("th")
        title = item.text
        print(title)


    results = soup.select('.mw-body-content')
    for item in results:
        a_item = item.select_one("th")
        title = item.text
        print(title)
    
get_all_href(url)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_html('https://en.wikipedia.org/wiki/Programming_languages_used_in_most_popular_websites')
df = data[0]
df
```

```python
websites = df['Websites']
uni_visitors = df['Popularity(unique visitors per month)[1]']
uni_visitors[4] = 516000000
uni_visitors[7] = 315000000

plt.pie(uni_visitors, 
        labels = websites, 
        radius =2,
        autopct = '%1.1f%%') #percentage
plt.show()
```
